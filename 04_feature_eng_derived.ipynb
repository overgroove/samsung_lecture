# **Derived feature**
- 데이터셋에 존재하는 변수를 활용하여 새로운 변수를 생성
- 기존 변수의 특징이 반영되지 않는 형태로 저장 되었을 경우
- 변수의 특징을 잘 설명할 수 있는(컴퓨터가 이해하기 편한) 파생변수 생성
- 데이터의 타입이나 특성/패턴 보다는 도메인에 따른 처리 필요

## **data loading**

```python
# 필요모듈 import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_openml

pd.options.display.max_columns = 200
pd.options.display.max_rows = 200
```

```python
# 데이터 로딩
data = fetch_openml('house_prices', as_frame=True)
df = data['frame']
df.head()
```

## **1. Generate simple calculate**

```python
# 주택 판매 당시 주택 연식 계산 'YrSold', 'YearBuilt'
# 시간데이터는 이후 발생한 시점 숫자가 항상 큽니다.
df.head()
df['Yrbe'] = df['YrSold'] - df['YearBuilt']
```

```python
# 판다스 구조
for i in [1, 2, 3, 4]:
    print(i + 1)
# 브로드캐스팅
df['YrSold'] + 1
```

```python
# 판매 전 마지막 리모델링 햇수 계산 'YrSold', 'YearRemodAdd'
df['YrRebe'] = df['YrSold'] - df['YearRemodAdd']
```

```python
# 총 주거면적 계산 'GrLivArea', 'TotalBsmtSF'
df['TotalLivAre'] = df['GrLivArea'] + df['TotalBsmtSF']
```

```python
# 차고 유무 'GarageArea'
(df['GarageArea'] == 0).sum()
df['Garage'] = df['GarageArea'] > 0
```

```python
# 총 욕실 갯수 'FullBath', 'HalfBath', 'BsmtFullBath', 'BsmtHalfBath'
df['BathArea'] = df['FullBath'] + (df['HalfBath'] * 2)  + df['BsmtFullBath'] + (df['BsmtHalfBath'] * 2)
```

```python
# 벽난로의 전체적인 품질 및 갯수를 아우르는 파생변수 생성
# 0. 결측데이터 처리
df['Fireplaces'].isna().any()
# 1. 벽난로 품질 변수 결측 데이터 처리
df['FireplaceQu'] = df['FireplaceQu'].fillna(0)
# 2. 품질 변수의 정수화
value_dict = {
    'Ex': 5,
    'Gd': 4,
    'TA': 3,
    'Fa': 2,
    'Po': 1,
    0: 0
}
# 3. 파생변수 연산 및 생성
df['FireplaceQu'] = df['FireplaceQu'].map(value_dict)
df['TotalFireplacesQu'] = df['Fireplaces'] * df['FireplaceQu']
```

- 변수의 특징을 잘 살릴 수 있는 파생변수가 생성 되었다면 해당 파생변수를 생성하기 위해 사용한 피연산 변수는 삭제해도 좋음
- 변수의 특징이 비슷하다면 해당 변수들을 한데 아우를 수 있는 파생변수를 만든다면 다중공선성 문제도 해결 가능함

## **2. Generate interaction**
- 단순 증감 연산이 아닌 관계형 변수도 생성 가능
- 합차연산이 아닌 곱연산을 통해 변수간 관계에 해당하는 파생변수도 생성 가능
- interaction 변수의 경우 기존 변수의 특징과 다른 새로운 특징을 가지기에 피연산 변수를 삭제 하지 않는 경우가 많음
- 다항 회귀에 주로 사용

```python
# 1층 주거면적의 quality OverallQual GrLivArea
# 결측데이터 확인 (결측없음)
(df['OverallQual'].isna().any())
(df['GrLivArea'].isna().any())
# 변수 값 확인(그대로 활용)
df['OverallQual']
# 연산적용 및 파생변수 생성
df['OverallQualGrLivArea'] = df['OverallQual'] * df['GrLivArea']
```

## **3. datetime 데이터 추출**

```python
# 시간속성 분해
df1['year'] = df1['datetime'].dt.year
df1['month'] = df1['datetime'].dt.month
df1['day'] = df1['datetime'].dt.day
df1['hour'] = df1['datetime'].dt.hour
df1['minute'] = df1['datetime'].dt.minute
df1['second'] = df1['datetime'].dt.second
df1['microsecond'] = df1['datetime'].dt.microsecond
df1['weekday'] = df1['datetime'].dt.dayofweek # 월요일부터 0 카테고리컬 화 인코딩 추천
df1
```

## **4. Binning feature**
- 수치형 변수의 연속성이 넓게 퍼진 분포를 보일 경우 값을 치환하는 변수 생성
- 나이 --> 연령대 처럼 수치적 rank 보다는 구간에 맞춰 특성의 구조를 변경
- 수치형 --> 범주형 변환

```python
# 주택 건설 년도 확인
df['YearBuilt'].min(), df['YearBuilt'].max()
```

```python
from pandas.core.reshape import encoding
# 구간화를 위한 sklearn.preprocessing KBinsDiscretizer
from sklearn.preprocessing import KBinsDiscretizer
kbins = KBinsDiscretizer(n_bins=10, encode='onehot-dense', strategy='quantile') # 몇구간? n_bins
kbins.fit(df[['YearBuilt']])
kbins.transform(df[['YearBuilt']])
```
